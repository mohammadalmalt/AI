```python
import os
import torch
import numpy as np
from diffusers import StableDiffusionPipeline
from moviepy.editor import ImageSequenceClip
from PIL import Image

# Check device: use GPU if available
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load a pre-trained text-to-image model (Stable Diffusion in this example)
# Note: Use a compatible version (e.g., stable-diffusion-v1-4)
print("Loading Stable Diffusion model...")
pipe = StableDiffusionPipeline.from_pretrained(
    "CompVis/stable-diffusion-v1-4",
    revision="fp16",
    torch_dtype=torch.float16
)
pipe = pipe.to(device)
print("Model loaded!")

# Define your text prompt and video parameters
prompt = "A mesmerizing futuristic cityscape at night, cinematic and ultra-detailed"
duration_seconds = 5        # total duration of the video
fps = 15                    # frames per second
num_frames = duration_seconds * fps

# Create a directory to store generated frames
frames_dir = "video_frames"
os.makedirs(frames_dir, exist_ok=True)

frames = []  # to hold frame images (as numpy arrays)
print("Generating video frames...")

for i in range(num_frames):
    # For a bit of variation, you can vary the random seed per frame.
    # For better temporal consistency, consider keeping the seed fixed or applying latent interpolation.
    seed = np.random.randint(0, 1000000)
    generator = torch.Generator(device=device).manual_seed(seed)

    # Generate an image from the prompt
    result = pipe(prompt, generator=generator)
    image: Image.Image = result.images[0]

    # Optionally, save each frame as an image file
    frame_filename = os.path.join(frames_dir, f"frame_{i:04d}.png")
    image.save(frame_filename)

    # Convert to a numpy array for MoviePy and append to our list of frames
    frames.append(np.array(image))

    print(f"Frame {i+1}/{num_frames} generated.")

print("All frames generated! Compiling frames into a video...")

# Use MoviePy to create a video clip from the frames
clip = ImageSequenceClip(frames, fps=fps)
output_video_path = "generated_video.mp4"
clip.write_videofile(output_video_path, codec="libx264")

print(f"Video saved as: {output_video_path}")
```

```python
import torch
from diffusers import StableVideoDiffusionPipeline
from IPython.display import Video

# Check environment and load model
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load pre-trained Stable Video Diffusion model
model_id = "google/ddpm-ema-celebahq-256"
pipe = StableVideoDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.to(device)

print("Stable Video Diffusion model loaded!")

# Define prompt and generation parameters
prompt = "A futuristic cityscape at night with neon lights"
num_frames = 16  # Number of frames to generate
frame_rate = 8  # Frames per second

# Generate video
print("Generating video...")
video_frames = pipe(prompt, num_frames=num_frames, frame_rate=frame_rate).frames

# Save the generated video
output_path = "generated_video.mp4"
video_frames.save(output_path)

print(f"High-quality video saved as: {output_path}")

# Display the video in a notebook (optional)
Video(output_path, embed=True)
```
