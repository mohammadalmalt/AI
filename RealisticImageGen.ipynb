```python
import torch
import shutil
from huggingface_hub import login
from diffusers import StableDiffusionPipeline, DPMSolverMultistepScheduler

# Step 1: Clear the cache for the model
cache_dir = "/root/.cache/huggingface/hub/models--SG161222--Realistic_Vision_V6.0_B1_noVAE"
try:
    shutil.rmtree(cache_dir)
    print(f"Cleared cache: {cache_dir}")
except FileNotFoundError:
    print(f"Cache directory not found: {cache_dir}")

# Step 2: Authenticate with Hugging Face Hub (if needed)
# Replace "YOUR_HUGGINGFACE_TOKEN" with your actual token
login(token="")

# Step 3: Load the Realistic Vision V6.0 B1 noVAE model
model_id = "SG161222/Realistic_Vision_V6.0_B1_noVAE"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)

# Disable the NSFW filter
pipe.safety_checker = None

# Step 4: Configure the scheduler (optional, but recommended for better quality)
pipe.scheduler = DPMSolverMultistepScheduler.from_config(
    pipe.scheduler.config,
    algorithm_type="dpmsolver++",  # Use "dpmsolver++" for better results
    final_sigmas_type="sigma_min",  # Avoids compatibility issues
)

# Step 5: Move model to GPU
pipe.to("cuda")

# Step 6: Define the prompt and negative prompt
prompt = "Medium shot portrait of a beautiful latin girl, highly detailed, photorealistic, 8k"
negative_prompt = "blurry, low quality, distorted, cartoonish, asymmetric face, deformed eyes, crooked nose, unnatural face"

# Step 7: Define the aspect ratio (e.g., 9:16 for portrait)
width = 768
height = 1024

# Step 8: Generate an image with controlled aspect ratio and more steps
image = pipe(
    prompt,
    negative_prompt=negative_prompt,
    width=width,
    height=height,
    num_inference_steps=250,  # Adjust for better quality
    guidance_scale=7.5,      # Adjust for creativity vs. adherence to prompt
).images[0]

# Step 9: Show and save the image
image.show()
image.save("generated_image_enhanced5.png")
```
